```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../data/titanic.csv',  index_col='PassengerId')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
new_df = df.drop('Cabin', axis=1).dropna()
```

```python
columns = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked', 'Survived']
sns.pairplot(new_df[columns]);
```

```python
sns.boxplot(x='Pclass', y='Fare', data=new_df);
```

```python
sns.countplot(x=new_df['Sex'], hue=new_df['Survived']);
```

```python
sns.countplot(x=new_df['Pclass'], hue=new_df['Survived']);
```

```python
new_df['Age_Group'] = 'Middle'
new_df.loc[new_df['Age'] < 30, 'Age_Group'] = 'Young'
new_df.loc[new_df['Age'] > 60, 'Age_Group'] = 'Old'
new_df.head()
```

```python
sns.countplot(x=new_df['Age_Group'], hue=new_df['Survived']);
```
