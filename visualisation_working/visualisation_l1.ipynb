```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option("display.max_columns", 1000)
pd.set_option("display.max_rows", 1000)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv("../data/Teen_Mental_Health_Dataset.csv")
df.head()
```

```python
### Количественные признаки
df['daily_social_media_hours'].hist();
```

```python
sns.boxplot(df['sleep_hours']);
```

```python
df.hist();
```

```python
### Категориальные признаки
sns.countplot(df['platform_usage']);
```

```python
sns.countplot(df['social_interaction_level']);
```

```python
### Взаимодействия
# Количественные-количественные
columns = ['daily_social_media_hours', 'sleep_hours', 'screen_time_before_sleep']
df[columns].hist();
```

```python
sns.pairplot(df[columns]);
```

```python
df['gender'].map({'male': 'blue', 'female': 'red'}).head()
```

```python
plt.scatter(df['sleep_hours'], df['daily_social_media_hours'],
            color=df['gender'].map({'male': 'blue', 'female': 'red'}),
            label='hours');
plt.xlabel('Sleep hours');
plt.ylabel('Daily social media hours');
plt.title('Sleep hours and social media hours for male/female');
plt.legend();
```

```python
plt.scatter(df[df['gender'] == 'male']['sleep_hours'], df[df['gender'] == 'male']['daily_social_media_hours'],
            color='blue', label='male hours');
plt.xlabel('Sleep hours');
plt.ylabel('Daily social media hours');
plt.legend();
```

```python
plt.scatter(df[df['gender'] == 'female']['sleep_hours'], df[df['gender'] == 'female']['daily_social_media_hours'],
            color='red', label='female hours');
plt.xlabel('Sleep hours');
plt.ylabel('Daily social media hours');
plt.legend();
```

```python
# Изменилось только визуально, но внутри также
# df.drop('stress_level', axis=1)
# Изменилось полностью
df.drop('stress_level', axis=1, inplace=True)
```

```python
# Количественно-категориальные и количественно-бинарный
sns.boxplot(x='platform_usage', y='screen_time_before_sleep', data=df);
plt.title('От чего труднее оторваться');
```

```python
sns.boxplot(x='gender', y='screen_time_before_sleep', data=df);
plt.title('Кто больше времени перед сном сидит в социальных сетях M/F');
```

```python
sns.countplot(x='platform_usage', hue='screen_time_before_sleep' ,data=df);
```
